```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load scores.csv
df = pd.read_csv('scores.csv')
```

```python
# plot normal and ul2 columns on the same graph
# with the epoch column as x-axis
# for each row connect the normal and ul2 value with a line

plt.plot(df['epoch'], df['normal'], label='normal')
plt.plot(df['epoch'], df['ul2'], label='ul2')
plt.xlabel('Steps')
plt.ylabel('Average Score Accuracy')
# set y scale to be between 0.34 and 0.35
plt.ylim(0.34, 0.35)
# connect the normal and ul2 value with a line
for i in range(len(df['epoch'])):
    plt.plot([df['epoch'][i], df['epoch'][i]], [df['normal'][i], df['ul2'][i]], color='red')

# add points for the normal and ul2 values
plt.scatter(df['epoch'], df['normal'], color='blue')
plt.scatter(df['epoch'], df['ul2'], color='orange')
plt.legend()
plt.show()

```
