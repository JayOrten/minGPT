```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "/home/jo288/nobackup/autodelete/cs674project2/runs/run3/loss.csv"
```

```python
df = pd.read_csv(csv_path)
```

```python
# FIrst column is iteration, secon column is loss
plt.plot(df.iloc[:, 0], df.iloc[:, 1])
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss vs Iteration")
plt.show()
```

```python
# Include line that is smoothed loss
smoothed_loss = df.iloc[:, 1].rolling(window=50).mean()
plt.plot(df.iloc[:, 0], smoothed_loss)
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss vs Iteration")
plt.show()
```

```python
# Plot both together as different colors
plt.plot(df.iloc[:, 0], df.iloc[:, 1], label="Loss")
plt.plot(df.iloc[:, 0], smoothed_loss, label="Smoothed Loss")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss vs Iteration")
plt.legend()
plt.show()
```
